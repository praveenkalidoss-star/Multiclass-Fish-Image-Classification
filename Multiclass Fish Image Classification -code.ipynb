# Step **2**

```python
BASE_PATH = "/content/data/images.cv_jzk6llhf18tm3k0kyttxz/data"
print(BASE_PATH)
```

```python
import tensorflow as tf
import os

train_data_dir = os.path.join(BASE_PATH, "train")
# Ensure the path is fully resolved and canonicalized for TensorFlow
train_data_dir = os.path.realpath(train_data_dir)

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_data_dir,
    image_size=(224, 224),
    batch_size=32,
    shuffle=True
)

print("Train classes:", train_ds.class_names)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    BASE_PATH + "/val",
    image_size=(224, 224),
    batch_size=32,
    shuffle=False
)

print("Val classes:", val_ds.class_names)
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    BASE_PATH + "/test",
    image_size=(224, 224),
    batch_size=32,
    shuffle=False
)

print("Test classes:", test_ds.class_names)
```

```python
for images, labels in train_ds.take(1):
    print("Image batch shape:", images.shape)
    print("Label batch shape:", labels.shape)
```

# Step **3**

```python
from tensorflow.keras import layers

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
])
```

```python
import matplotlib.pyplot as plt

for images, labels in train_ds.take(1):
    augmented_images = data_augmentation(images)

    plt.figure(figsize=(8, 8))
    for i in range(9):
        ax = plt.subplot(3, 3, i+1)
        plt.imshow(augmented_images[i].numpy().astype("uint8"))
        plt.axis("off")
    break
```

# step **4**

```python
from tensorflow.keras import layers, models

class_names = train_ds.class_names

cnn_model = models.Sequential([

    # Data augmentation (optional but helpful)
    data_augmentation,

    # 1st Convolution Block
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2,2)),

    # 2nd Convolution Block
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    # 3rd Convolution Block
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    # Flattening
    layers.Flatten(),

    # Fully Connected Layers
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),     # reduce overfitting

    # Output Layer
    layers.Dense(len(class_names), activation='softmax')
])

cnn_model.summary()
```

```python
cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = cnn_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.legend()
plt.title("Accuracy")
plt.show()

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title("Loss")
plt.show()
```

# step **5**

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models
import tensorflow as tf
```

```python
vgg = VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)

vgg.trainable = False  # Freeze VGG16 layers
```

```python
vgg_model = models.Sequential([
    data_augmentation,           # from step 3
    layers.Rescaling(1./255),    # normalize pixel values
    vgg,                         # VGG16 base model
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(len(class_names), activation='softmax')
])

vgg_model.summary()
```

```python
vgg_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history_vgg = vgg_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

```python
import matplotlib.pyplot as plt

# Accuracy
plt.figure(figsize=(6,4))
plt.plot(history_vgg.history['accuracy'], label='Train Acc')
plt.plot(history_vgg.history['val_accuracy'], label='Val Acc')
plt.title("VGG16 Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Loss
plt.figure(figsize=(6,4))
plt.plot(history_vgg.history['loss'], label='Train Loss')
plt.plot(history_vgg.history['val_loss'], label='Val Loss')
plt.title("VGG16 Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

# Model 2 **RESNET50**

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models
import tensorflow as tf

resnet_base = ResNet50(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)

resnet_base.trainable = False  # freeze weights
```

```python
resnet_model = models.Sequential([
    data_augmentation,
    layers.Rescaling(1./255),
    resnet_base,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(len(class_names), activation='softmax')
])

resnet_model.summary()
```

```python
resnet_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history_resnet = resnet_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```

```python
import matplotlib.pyplot as plt

# Accuracy
plt.plot(history_resnet.history['accuracy'], label='Train Acc')
plt.plot(history_resnet.history['val_accuracy'], label='Val Acc')
plt.title("ResNet50 Accuracy")
plt.legend()
plt.show()

# Loss
plt.plot(history_resnet.history['loss'], label='Train Loss')
plt.plot(history_resnet.history['val_loss'], label='Val Loss')
plt.title("ResNet50 Loss")
plt.legend()
plt.show()
```
